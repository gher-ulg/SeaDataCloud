Goal: compare the fields from two netCDF before and after the re-gridding operation

```python
import os
import netCDF4
import matplotlib.pyplot as plt
```

```python
datadir = "../data/"
datafile1 = os.path.join(datadir, "input_latlon.nc")
datafile2 = os.path.join(datadir, "output_fine.nc")
```

```python
with netCDF4.Dataset(datafile1, 'r') as nc:
    lon1 = nc.variables['lon'][:]
    lat1 = nc.variables['lat'][:]
    field1 = nc.variables['analyzed_field'][:]
with netCDF4.Dataset(datafile2, 'r') as nc:
    lon2 = nc.variables['lon'][:]
    lat2 = nc.variables['lat'][:]
    field2 = nc.variables['analyzed_field'][:]
```

```python
fig = plt.figure(figsize=(10, 8))
ax1 = plt.subplot(211)
ax1.set_xticks([])
ax1.set_yticks([])
ax1.axis('off')
plt.title("Product 1", fontsize=22)
plt.pcolormesh(lon1, lat1, field1)
ax2 = plt.subplot(212)
ax2.set_xticks([])
ax2.set_yticks([])
ax2.axis('off')
plt.title("Product 2", fontsize=22)
plt.pcolormesh(lon2, lat2, field2)
plt.savefig("regrid_comparison", dpi=300, bbox_inches="tight")
plt.show()
plt.close()
```
