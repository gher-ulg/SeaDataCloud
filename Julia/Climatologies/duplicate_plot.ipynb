```python
import os
import matplotlib.pyplot as plt
import netCDF4
import numpy as np
```

```python
datadir = "/data/SeaDataCloud/NorthSea/"
obsfile = os.path.join(datadir, "NorthSea_obs.nc")
obsfile_wod = os.path.join(datadir, "NorthSea_obs_wod.nc")
```

```python
with netCDF4.Dataset(obsfile) as nc1:
    lon1 = nc1.variables["obslon"][:]
    lat1 = nc1.variables["obslat"][:]
with netCDF4.Dataset(obsfile_wod) as nc2:
    lon2 = nc2.variables["obslon"][:]
    lat2 = nc2.variables["obslat"][:]
print("Dataset no.1: {}".format(len(lon1)))
print("Dataset no.2: {}".format(len(lon2)))
```

```python
sel = np.where(lon1 > 180.)[0]
lon1[sel] = lon1[sel] - 360.
sel = np.where(lon2 > 180.)[0]
lon2[sel] = lon2[sel] - 360.
```

```python
fig = plt.figure()
plt.plot(lon1, lat1, "ko", ms=.1)
plt.plot(lon2, lat2, "ro", ms=.1)
plt.xlim(0.,1.)
plt.ylim(53., 54.)
plt.show()
plt.close()
```
